```python
from PIL import Image, ImageEnhance

# Open the image
image_path = "/top2.jpeg"
obrazek = Image.open(image_path)

# Apply saturation enhancement
enhancer = ImageEnhance.Color(obrazek)
obrazek = enhancer.enhance(1.5)  # Increase saturation by a factor of 1.5

# Apply contrast enhancement
enhancer = ImageEnhance.Contrast(obrazek)
obrazek = enhancer.enhance(1.2)  # Increase contrast by a factor of 1.2

# Display the modified image

display(obrazek)

```
