```python
from PIL import Image, ImageEnhance, ImageFilter



def saturace():
  obrazek = Image.open("/content/top2.jpeg")

  enhancer = ImageEnhance.Color(obrazek)
  obrazek = enhancer.enhance(1.5)

  enhancer = ImageEnhance.Contrast(obrazek)
  obrazek = enhancer.enhance(1.2)
  return(obrazek)
def cernobila():
  obrazek = Image.open("/content/top2.jpeg")
  sirka, vyska = obrazek.size

  x = 0
  while x < sirka:
      y = 0
      while y < vyska:
          r, g, b = obrazek.getpixel((x, y))

          grayscale = int((r + g + b) / 3)

          obrazek.putpixel((x, y), (grayscale, grayscale, grayscale))

          y += 1
      x += 1
  return(obrazek)
def rozmazani():
  obrazek = Image.open("/content/top2.jpeg")
  return obrazek.filter(ImageFilter.BLUR)
while True:
    print("\n----- editors Database -----")
    print("1. filtr 1")
    print("2. filtr 2")
    print("3. filtr 3")
    print("4. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        modified_image = saturace()

    elif choice == "2":
        modified_image = cernobila()
    elif choice == "3":
        modified_image = rozmazani()
    elif choice == "4":
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a corect option [1,2,3].")
    display(modified_image)
```
